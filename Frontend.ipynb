```python
from joblib import load
import pandas as pd
import tkinter as tk
from tkinter import ttk, messagebox
```

```python
model = load("heart_disease_model.joblib")
feature_names = load("feature_names.joblib")
```

```python
def predict_result():
    try:
        sex_map = {"Male": 1, "Female": 0}
        bin_map = {"Yes": 1, "No": 0}

        data = {
            "age": int(age_var.get().strip()),
            "sex": sex_map[sex_var.get()],
            "trestbps": float(trestbps_var.get().strip()),
            "chol": float(chol_var.get().strip()),
            "fbs": bin_map[fbs_var.get()],
            "thalch": float(thalch_var.get().strip()),
            "exang": bin_map[exang_var.get()],
            "oldpeak": float(oldpeak_var.get().strip()),
            "fbs_missing": 0,
            "exang_missing": 0,
            "slope_downsloping": 0,
            "slope_flat": 0,
            "slope_upsloping": 0,
            "cp_atypical angina": 0,
            "cp_non-anginal": 0,
            "cp_typical angina": 0,
            "restecg_normal": 0,
            "restecg_st-t abnormality": 0
        }

        slope_value = slope_var.get().strip().lower()
        cp_value = cp_var.get().strip().lower()
        restecg_value = restecg_var.get().strip().lower()

        if slope_value == "downsloping":
            data["slope_downsloping"] = 1
        elif slope_value == "flat":
            data["slope_flat"] = 1
        elif slope_value == "upsloping":
            data["slope_upsloping"] = 1

        if cp_value == "atypical angina":
            data["cp_atypical angina"] = 1
        elif cp_value == "non-anginal":
            data["cp_non-anginal"] = 1
        elif cp_value == "typical angina":
            data["cp_typical angina"] = 1

        if restecg_value == "normal":
            data["restecg_normal"] = 1
        elif restecg_value == "st-t abnormality":
            data["restecg_st-t abnormality"] = 1

        df = pd.DataFrame([data])
        df = df[feature_names]

        pred = model.predict(df)[0]

        print(df)
        print("Prediction:", pred)

        if hasattr(model, "predict_proba"):
            print("Probabilities:", model.predict_proba(df))

        if pred == 0:
            result_var.set("No Heart Disease")
            result_label.config(fg="green")
        else:
            result_var.set("Heart Disease Detected")
            result_label.config(fg="red")

    except Exception as e:
        messagebox.showerror("Error", str(e))
```

```python
root = tk.Tk()
root.title("Heart Disease Prediction")
root.geometry("520x700")

root.grid_columnconfigure(0, weight=1)
root.grid_columnconfigure(1, weight=1)
root.grid_columnconfigure(2, weight=1)

age_var = tk.StringVar()
sex_var = tk.StringVar(value="Male")
trestbps_var = tk.StringVar()
chol_var = tk.StringVar()
fbs_var = tk.StringVar(value="No")
thalch_var = tk.StringVar()
exang_var = tk.StringVar(value="No")
oldpeak_var = tk.StringVar()
slope_var = tk.StringVar(value="Flat")
cp_var = tk.StringVar(value="Typical Angina")
restecg_var = tk.StringVar(value="Normal")
result_var = tk.StringVar()

tk.Label(root, text="Heart Disease Prediction", font=("Arial", 16, "bold")).grid(row=0, column=0, columnspan=3, pady=20)

fields = [
    ("Age", age_var),
    ("Resting BP", trestbps_var),
    ("Cholesterol", chol_var),
    ("Max Heart Rate", thalch_var),
    ("Oldpeak", oldpeak_var)
]

for i, (label, var) in enumerate(fields, start=1):
    tk.Label(root, text=label).grid(row=i, column=0, padx=10, pady=6, sticky="e")
    tk.Entry(root, textvariable=var, width=25).grid(row=i, column=1, padx=10, pady=6)

tk.Label(root, text="Sex").grid(row=6, column=0, padx=10, pady=6, sticky="e")
ttk.Combobox(root, textvariable=sex_var, values=["Male", "Female"], state="readonly", width=22).grid(row=6, column=1, pady=6)

tk.Label(root, text="FBS").grid(row=7, column=0, padx=10, pady=6, sticky="e")
ttk.Combobox(root, textvariable=fbs_var, values=["Yes", "No"], state="readonly", width=22).grid(row=7, column=1, pady=6)

tk.Label(root, text="Exercise Angina").grid(row=8, column=0, padx=10, pady=6, sticky="e")
ttk.Combobox(root, textvariable=exang_var, values=["Yes", "No"], state="readonly", width=22).grid(row=8, column=1, pady=6)

tk.Label(root, text="Slope").grid(row=9, column=0, padx=10, pady=6, sticky="e")
ttk.Combobox(root, textvariable=slope_var, values=["Downsloping", "Flat", "Upsloping"], state="readonly", width=22).grid(row=9, column=1, pady=6)

tk.Label(root, text="Chest Pain").grid(row=10, column=0, padx=10, pady=6, sticky="e")
ttk.Combobox(root, textvariable=cp_var, values=["Atypical Angina", "Non-Anginal", "Typical Angina"], state="readonly", width=22).grid(row=10, column=1, pady=6)

tk.Label(root, text="Rest ECG").grid(row=11, column=0, padx=10, pady=6, sticky="e")
ttk.Combobox(root, textvariable=restecg_var, values=["Normal", "ST-T Abnormality"], state="readonly", width=22).grid(row=11, column=1, pady=6)

tk.Button(root, text="Predict", command=predict_result, width=20).grid(row=12, column=0, columnspan=3, pady=20)

result_label = tk.Label(root, textvariable=result_var, font=("Arial", 14, "bold"))
result_label.grid(row=13, column=0, columnspan=3, pady=20)

root.mainloop()
```
